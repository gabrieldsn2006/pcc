```python
from digraph import Digraph
from DirectedEulerianCycle import DirectedEulerianCycle
```

### Construção do Dígrafo Ponderado

```python
def load_graph_from_file(file_path):
    with open(file_path, 'rt') as f:
        v = int(f.readline())
        e = int(f.readline())
        G = Digraph(v)
        for line in f.readlines():
            parts = line.split()
            G.add_edge(int(parts[0]), int(parts[1]), int(parts[2]))
    return G

G = load_graph_from_file('../dados/entrada_eulerizada.txt')

print(f'Lista de Adjacência:\n{G}')
```

### Graus de Entrada e Saída dos Vértices
grau de entrada igual ao grau de saída identifica um vértice balanceado.

```python
for v in range(G.V):
    print(f'{v.__str__().replace("0", "a").replace("1", "b").replace("2", "c").replace("3", "d").replace("4", "e").replace("5", "f")}: in: {G.in_degree(v)} - out: {G.out_degree(v)}')
```

### Encontrando Circuito Euleriano a partir do Método de Hierholzer

```python
# Testar o algoritmo de Hierholzer
h = DirectedEulerianCycle(G)

print(f'Tem circuito Euleriano: {h.has_circuit()}')
print(f'\n{h}')
print(f'Custo total do circuito: {h.circuit_cost()}')

if h.has_circuit():
    print(f'\nCircuito: {h.get_circuit().__str__().replace("0", "a").replace("1", "b").replace("2", "c").replace("3", "d").replace("4", "e").replace("5", "f")}')
```
