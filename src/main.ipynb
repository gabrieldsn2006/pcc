```python
from digraph import Digraph
```

```python
def load_graph_from_file(file_path):
    with open(file_path, 'rt') as f:
        v = int(f.readline())
        e = int(f.readline())
        G = Digraph(v)
        for line in f.readlines():
            parts = line.split()
            G.add_edge(int(parts[0]), int(parts[1]))
    return G

G = load_graph_from_file('../dados/entrada_original.txt')

print(f'Lista de Adjacência:\n{G}')
```
